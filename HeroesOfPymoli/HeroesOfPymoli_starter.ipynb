## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
Agedem = purchase_data.loc[:,["Age", "SN"]]
players = Agedem["SN"].nunique()
playercnt = Agedem["SN"].count()
percentplayers = (playercnt/players)

bins = [0,10,15,20,25,30,35,40,100]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

Agedem["Age Group"] = pd.cut(Agedem["Age"], bins, labels=age_labels)
Agegrp2 = Agedem["Age Group"]
Agedem["Percentage of Players"]= percentplayers
Agedem["Total Count"] = playercnt

AgedemSm = {"Age Group": (Agegrp2),"Total Count":(playercnt),"Percentage of Players":(percentplayers)}
df_Agedemsm = pd.DataFrame(AgedemSm)
df_Agedemsm.set_index('Age Group', inplace=True)
#df_Agedemsm.head(10)
Agedemgroup = df_Agedemsm.groupby(['Age Group'])
Agedemgroup.count().head(10)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
PurchAge = purchase_data.loc[:,["Age", "Item ID", "Price"]]

Purchases_A = PurchAge["Item ID"].count()
avg_price_A = PurchAge["Price"].mean()
tot_purch_valA = PurchAge["Price"].sum()
avg_purchper_person= (tot_purch_valA/Purchases_A)


bins = [0,10,15,20,25,30,35,40,100]
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
PurchAge["Age Ranges"] = pd.cut(PurchAge["Age"], bins, labels=age_labels)
Agerange2 = PurchAge["Age Ranges"]

PurchA = {"Age Ranges":(Agerange2),
    "Purchase Count":(Purchases_A), 
    "Average Purchase Price":(avg_price_A), 
    "Total Purchase Value":(tot_purch_valA),
    "Avg Total Purchase per Person":(avg_purchper_person)}

df_PurchA = pd.DataFrame(PurchA)
df_PurchA.set_index("Age Ranges", inplace=True)
df_PurchA.head()
#PurchAge.head()
PurchAgroup = df_PurchA.groupby(["Age Ranges"])
PurchAgroup.count().head()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting

```python
TpSpnd = purchase_data.loc[:, ["SN", "Price"]]

TotPurchVal = TpSpnd["Price"].sum()
Purchcount = TpSpnd["SN"].value_counts()

df_Purchcount = pd.DataFrame(Purchcount)
df_Purchcount_Rename = df_Purchcount.rename(columns={"SN": "Purchase Count"})

AvgPurchPrice = (TotPurchVal/df_Purchcount_Rename["Purchase Count"])
df_Purchcount_Rename["Average Purchase Price"] = (AvgPurchPrice)/(Purchcount)
df_Purchcount_Rename["Total Purchase Value"] = TotPurchVal

df_Purchcount_Rename["Average Purchase Price"]= df_Purchcount_Rename["Average Purchase Price"].astype(float).map("${:,.2f}".format)
df_Purchcount_Rename["Total Purchase Value"]= df_Purchcount_Rename["Total Purchase Value"].astype(float).map("${:,.2f}".format)
df_Purchcount_Rename.head()
```
